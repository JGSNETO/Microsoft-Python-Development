# Testing Basics & Version Control 

## Why unit tests matter

- In python a unit test is a piece of code that tests a speciic function or methos in your program. 
- *Why unit testing is essential*:
1. Catching errors early 
2. Improve code quality
3. Making changes less risky 

- *When to use unit testing*:
1. Complex projects 
2. Collaborative environments 
3. Long-term maintenance 

## Unit testing: Removing bugs from your code

- *What is unit testing?*:
1. Dissect your code
2. Examine each component

- *The unit testing process*
1. Identify the unit
2. Define the test case
3. Write test code 
4. Execute the tests
