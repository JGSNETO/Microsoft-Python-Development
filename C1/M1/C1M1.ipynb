## Programming Fundamentals 

- The simplest definition of programming is that it's the way people communicate with computers and instruct them to perform tasks.

- Programming fundamentals:
1. Variables: Containers to store data. 
2. Data types: Define the different type of data that variables can hold. 
3. Operators: Common operators(Arithmetic, Logical)
4. Control flow statements: Determine the order the code is executed. 
5. Functions: Reusable blocks of code that perform specific tasks, breaking down complex actions in small and more simple activities. 

## Explaining Python

- Python defining characteristics:
1. *Interpreted language*: Python code is executed, line by line, by an interpreter. This is different from compiled languages like C++ or Java, where the entire code is converted to machine code before it is executed. You can run and see the result directly, without a extra compilation step. 
2. *High-level language*: Focus on the logic of your program, rather than the details of how the computer executes it. 
3. *Object-Oriented Programming (OOP)*: Paradigm that organizes code around data structures that contain both data and functions. Which are called objects. Objects promote code reusability, modularity, and maintainability, making it easier to manage and scale large software projects.
