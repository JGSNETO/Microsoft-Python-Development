# Anatomy of a Python Program

Python, celebrated for its simple syntax and remarkable versatility, has cemented its place as a fundamental tool in the software development landscape. ## Statements: The commands that orchestrate your code´s symphony
Imagine you're guiding an orchestra. You use specific gestures and cues to direct each musician, creating a harmonious symphony. In Python, these cues are comparable to statements. The individual lines of code serve as instructions, dictating the actions your computer should perform.

## Assignment statement 
The line age = 25 assigns the numerical value 25 to the variable age. This is like setting up a character in a play, establishing their initial state. Variables act as containers for storing data, allowing you to reference and manipulate that data throughout your program. In this case, the variable age holds the age of the person we're evaluating.

## Conditional statements and comparison operators
The if age >= 18: and else: statements introduce decision points. The computer evaluates the condition (age >= 18) and then executes the corresponding block of code based on whether the condition is true or false. This is how we create branching paths in our programs, much like a choose-your-own-adventure story. Conditional statements allow us to adapt the behavior of our code based on specific circumstances. For instance, we only print the "You are eligible to vote" message in our example if the age variable meets the minimum voting age requirement.

The core of these conditions lies with comparison operators. In our example, >= (greater than or equal to) is a comparison operator that checks the relationship between age and 18. Other common comparison operators include == (equal to), != (not equal to), > (greater than), and < (less than). The != operator, specifically, checks if two values are not the same, such as if user_input != "quit": to continue a program until a specific word is typed. These operators are fundamental because they produce a Boolean result (either True or False), which is exactly what conditional statements need to decide which path to take.

It's crucial to remember that a single equals sign (=) is used for assignment (giving a variable a value, like age = 25), while a double equals sign (==) is used for comparison (checking if two values are the same). 

## Print statements
The lines print("You are eligible to vote.") and print("You are not yet eligible to vote.") display messages to the user. These are our way of communicating the results of our computations or providing feedback to the user. Print statements are essential for interacting with users, debugging code, and presenting information in a readable format.

```python
age = 25
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not yet eligible to vote.")
```

## Less frequent arithmetic operations
Floor Division // Performs division and discards any remainder, resulting in a whole number (so 7 // 4 is 1, as 4 goes into 7 a single time evenly). May also be referred to as integer division.

Modulo % Returns the remainder after integer division (so 7 % 4 is 3, as the division has a remainder of 3).

Exponentiation ** Raises a number to the power of another number. To write 24, or two to the fourth power, you write 2 ** 4.

```python
discount = 0.2  # 20% discount
price = 50
final_price = price * (1 - discount)
print(final_price)
```

In this code snippet:

- 0.2 and 50 are simple expressions that represent numerical values. These values can be directly used in calculations or stored in variables for later use.

- 1 - discount is a more complex expression that calculates the discount percentage (for example, a discount of 20% means the price is 80% of the original cost). This demonstrates how expressions can be combined using operators like subtraction. The parentheses are used to control the order of operations, ensuring that the discount is subtracted from 1 before the multiplication takes place.

- price * (1 - discount)  is the final expression that determines the final_price by multiplying the original price by the calculated discount. This showcases the versatility of expressions, allowing us to perform arithmetic operations and manipulate variables to get meaningful results.

Expressions are the gears and levers that drive the inner workings of your programs. They allow you to manipulate data, make comparisons, and find new information. From simple arithmetic to complex string manipulations, expressions are the tools that enable your code to reason and adapt.

Imagine you're building a program to calculate the area of a circle. The ** operator is used for exponentiation, squaring the radius.

## Basic Automation
*Iterate*: Systematically loop through all the data files, ensuring that every data point is included in the analysis. Loops, such as for loops, allow you to repeat a block of code for each item in a collection, such as a list of files. This eliminates the need for manual repetition and ensures consistency in your analysis.

*Extract*: Intelligently read the relevant data from each file, filtering out noise and irrelevant information. Python provides libraries like pandas, which simplifies the process of reading data from various file formats (for example, CSV or Excel) and manipulating it in a structured manner. You can easily select specific columns, filter rows based on conditions, and extract the data you need for your analysis.

Here is a simple CSV file example:

```python
import csv

# Read a CSV file
with open('data.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        print(row)
```
